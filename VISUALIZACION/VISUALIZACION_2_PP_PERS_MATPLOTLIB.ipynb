```python
plt.subplot(121)
plt.rcParams['figure.figsize'] = (20, 10)
plt.xlabel("NUMERO ATAQUES")
plt.ylabel("TIPO ATAQUE")
plt.title('ESTADOS UNIDOS')
plt.plot(ataques_usa, label= 'ataques_usa', color= 'red', marker = "o", markersize = 7)
plt.legend

plt.subplot(122)
plt.rcParams['figure.figsize'] = (20, 10)
plt.xlabel("NUMERO ATAQUES")
plt.ylabel("TIPO ATAQUE")
plt.title('ESPAÑA')
plt.plot(ataques_españa, label= 'ataques_españa', color= 'green', marker = "o", markersize = 12)
plt.legend
```
