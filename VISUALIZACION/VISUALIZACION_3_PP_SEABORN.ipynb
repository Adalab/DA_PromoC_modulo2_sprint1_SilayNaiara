1. Usando la librería seaborn, crea una gráfica de barras que muestre el número de personas que han hecho cada actividad en cada año. Hacedlo solo desde el 2015. Os podrá salir una gráfica un poco fea, no os preocupeis, es lo que esperamos.

```python
ataques_panda = pd.read_csv("../PANDAS/ataques_pandas7.csv", index_col=0)
ataques_panda.head()
```

```python
ataques_panda.columns
```

```python
ataques_desde_2015 = ataques_panda[ataques_panda["YEAR"] >= 2015]
ataques_desde_2015.head()
```

```python
plt.figure(figsize =(8,10))

# elegimos el método hisplot con la variable que queremos visualizar
fig = sns.histplot(x=ataques_desde_2015["YEAR"], y=ataques_desde_2015["ACTIVITY"], data = ataques_panda)

# establecemos el nombre de los ejes, pero recordad que esto no es obligatorio ya que Seaborn nos los genera automaticamente
fig.set(xlabel='AÑOS', ylabel = 'ACTIVIDAD')

# Ponemos título a  la gráfica
fig.set_title('Ataques/Actividad desde 2015' )

# mostramos la gráfica
plt.show()
```

4. Seleccionad solo los registros de USA, AUSTRALIA y SOUTH AFRICA. Muestra el número de ataques por país y año desde el año 1900.

```python
ataques_paises = ataques_panda[(ataques_panda["COUNTRY"] == "USA") | (ataques_panda["COUNTRY"] == "AUSTRALIA") | (ataques_panda["COUNTRY"] == "SOUTH AFRICA")]
ataques_paises.head()
```

```python
ataques_desde_1900 = ataques_paises[ataques_paises["YEAR"] >= 1900]
ataques_desde_1900.head()
```

```python
plt.figure(figsize =(20,10))

# elegimos el método hisplot con la variable que queremos visualizar
fig = sns.histplot(x=ataques_desde_1900["YEAR"], data = ataques_desde_1900, hue=ataques_desde_1900["COUNTRY"])

# Ponemos título a  la gráfica
fig.set_title('Ataques/pais desde 1900' )

# mostramos la gráfica
plt.show()
```
