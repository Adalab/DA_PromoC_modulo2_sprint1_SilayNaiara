```python
# Utilizando métodos de pandas: media, mediana, moda, máximo y mínimo.
# Media.
media = tik_tok["artist_pop"].mean()
media
```

```python
# Mediana.
mediana = tik_tok["artist_pop"].median()
mediana
```

```python
# Moda.
moda = tik_tok["artist_pop"].mode()
moda
```

```python
# Máximo
maximo = tik_tok["artist_pop"].max()
maximo
```

```python
# Mínimo
minimo = tik_tok["artist_pop"].min()
minimo
```

```python
# Percentil 25.
percentil25 = np.percentile(tik_tok["artist_pop"], 25)
print("El percentil 25 es:", percentil25)
```

```python
# Percentil 75.
percentil75 = np.percentile(tik_tok["artist_pop"], 75)
print("El percentil 75 es:", percentil75)
```

```python
# Rango Intercualtílico.

# Calculamos los percentiles 75 y 25 (que serían el tercer y primer cuartil)
q3, q1 = np.percentile(tik_tok["artist_pop"], [75, 25])

# Calculamos el rango intercuartílico como la resta de los valores del tercer y primer cuartil
riq = q3 - q1
print("El rango intercuartílico es:", riq)
```

```python
# Boxplot de algunas variables.
# Establecemos el tamaño de la gráfica
fig = plt.figure(figsize = (10,5))

# creamos el boxplot
plt.boxplot(tik_tok["artist_pop"]);
```

```python
# Tabla de frecuencias.
# Primero agrupamos el dataframe por la columna artist_pop
tabla_frec = tik_tok.groupby("artist_pop").count().reset_index() # Devolvera un objeto DataFrameGroupBy
tabla_frec.head()
```

```python
# Variables cualitativas.
var_cual = tik_tok.groupby("artist_pop").count().reset_index() # Devolvera un objeto DataFrameGroupBy
var_cual.head()
```

```python
# Variables cuantitativas.
var_cuant = tik_tok.groupby("artist_pop").count().reset_index().sort_values(by = "danceability", ascending = False)
var_cuant.head()
```

```python
# Frecuencia acumulada.
frec_acumulada = np.add(var_cual, var_cuant)
frec_acumulada.head()
```
